```python
import pandas

pandas.options.display.max_columns = None
pandas.options.display.max_rows = None

pokemon_dataset = pandas.read_csv("pokemon.csv")
'''Which pokemon has the maximum attack amongst the pokemon from 1ST generation'''
max_attack = pokemon_dataset[pokemon_dataset["generation"] == 1].sort_values("attack", ascending = False)
max_attack.head(10)
```

```python
'''Which legendary pokemon has the highest speed?'''
max_speed = pokemon_dataset[pokemon_dataset["is_legendary"] == 1].sort_values("speed", ascending = False)
max_speed.head(10)
```

```python
'''Which pokemon from generation 5 has the highest defence?'''
max_defence = pokemon_dataset[pokemon_dataset["generation"] == 5].sort_values("defense", ascending = False)
max_defence.head(10)
```

```python
'''Which pokemon has the lowest hp?'''
max_hp = pokemon_dataset.groupby (["name","classfication", "type1"])["hp"].mean().sort_values()
max_hp.head(10)
```
